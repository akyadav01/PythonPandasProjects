```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

```python
forbes_df = pd.read_csv('2022_forbes_billionaires.csv')
```

```python
## DROP FIRST COLUMN FROM DATA FRAME
forbes_df.drop(columns=forbes_df.columns[0], axis=1,  inplace=True)
forbes_df.head(3)
```

```python
## TOP 10 BILLIONAIRES LIST
fig = plt.figure(figsize=(10, 5))
plt.bar(forbes_df['name'][: 10], forbes_df['networth'][: 10], color ='blue', 
        width = 0.4)
 
plt.xlabel("Billionaire Name")
plt.ylabel("Billionaire's Networth")
plt.title("Top 10 Billionaires Name and Networth")
plt.show()
```

```python
## Countries with highest number of Forbes list companies

count_industries = forbes_df.groupby(['country'])['industry'].count().sort_values(ascending=False)
count_industries
```

```python
## number of billionaires in the country

count_billionaires = forbes_df.groupby(['country'])['name'].count().sort_values(ascending=False)
count_billionaires
```
